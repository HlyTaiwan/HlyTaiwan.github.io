```python
test_df = pd.ExcelFile('raw_data/Aj_s wine _ bistro.xlsx').parse(0)
test_df = test_df.dropna(subset='d4r55').reset_index(drop=True)
#https://maps.gstatic.com/consumer/images/icons/2x/ic_star_rate_14.png
#https://maps.gstatic.com/consumer/images/icons/2x/ic_star_rate_empty_14.png
test_list = []
for index_test in range(len(test_df)):
  test_count_star = []
  test_count_star = [test_df['hCCjke src 5'][index_test],test_df['hCCjke src 4'][index_test],test_df['hCCjke src 3'][index_test],test_df['hCCjke src 2'][index_test],test_df['hCCjke src'][index_test]]
  test_list.append(test_count_star.count('https://maps.gstatic.com/consumer/images/icons/2x/ic_star_rate_14.png'))

for index_test_1 in range(len(test_df)):
  if re.search('[\u4e00-\u9fa5]',test_df['wiI7pd'][index_test_1]) == None:
    print(index_test_1)
```

```python
star = 'https://maps.gstatic.com/consumer/images/icons/2x/ic_star_rate_14.png'
enptystar = 'https://maps.gstatic.com/consumer/images/icons/2x/ic_star_rate_empty_14.png'

data_reviews,rest_name,user_name,rate,review_time,review,feedback=[],[],[],[],[],[],[]
for restName in restNames['name']:
  df = pd.ExcelFile(f'raw_data/{restName}.xlsx').parse(0)
  df = df.dropna(subset='d4r55').reset_index(drop=True)
  data_reviews.append(len(df))

  for row in range(len(df)):
    count_star = []
    rest_name.append(restName)
    user_name.append(df['d4r55'][row])
    count_star = [df['hCCjke src 5'][row],df['hCCjke src 4'][row],df['hCCjke src 3'][row],df['hCCjke src 2'][row],df['hCCjke src'][row]]
    rate.append(count_star.count(star))
    review_time.append(df['rsqaWe'][row])
    review.append(df['wiI7pd'][row])
    try:
      feedback.append(df['wiI7pd 2'][row])
    except KeyError:
      feedback.append(None)
```

```python
rest_review = pd.DataFrame({
    'rest_name':rest_name,
    'user_name':user_name,
    'rate':rate,
    'review_time':review_time,
    'review':review,
    'feedback':feedback
})
#rest_review.to_excel('rest_review.xlsx', index = False)
```

```python
non_ENG = []
zhmodel = re.compile(u'[\u4e00-\u9fa5]')
for x in range(len(review)):
  if zhmodel.search(u'{}'.format(review[x])) != None:
    non_ENG.append(True)
  elif review[x] == '':
    non_ENG.append(True)
  else:
    non_ENG.append(False)
```

```python
rest_review_nonENG = rest_review[non_ENG]
#rest_review_nonENG.to_excel('rest_review_nonENG.xlsx', index = False)
```

```python
after_delete = rest_review_nonENG.drop_duplicates(subset=['rest_name'])['rest_name'].tolist()
original_name = restNames['name'].tolist()
for x in after_delete:
  original_name.remove(x)
```

```python
outlier = restNames.index[restNames['name'] == '左巴好室．好事'].tolist()
restNames.drop(outlier[0], inplace = True)
restNames['data_reviews'] = rest_review_nonENG['rest_name'].value_counts(sort=False).tolist()
restNames['data_rate'] = rest_review_nonENG.groupby('rest_name', sort = False)['rate'].mean().tolist()

restNames.to_csv('restaurant.csv', index = False)
```
